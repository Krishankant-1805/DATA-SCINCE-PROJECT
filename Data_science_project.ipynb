### What is numpy?

NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.


At the core of the NumPy package, is the ndarray object. This encapsulates n-dimensional arrays of homogeneous data types

### Numpy Arrays Vs Python Sequences

- NumPy arrays have a fixed size at creation, unlike Python lists (which can grow dynamically). Changing the size of an ndarray will create a new array and delete the original.

- The elements in a NumPy array are all required to be of the same data type, and thus will be the same size in memory.

- NumPy arrays facilitate advanced mathematical and other types of operations on large numbers of data. Typically, such operations are executed more efficiently and with less code than is possible using Python’s built-in sequences.

- A growing plethora of scientific and mathematical Python-based packages are using NumPy arrays; though these typically support Python-sequence input, they convert such input to NumPy arrays prior to processing, and they often output NumPy arrays.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Online Dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv"

# Load Dataset
data = pd.read_csv(url)

print("Dataset Shape:", data.shape)
print(data.head())

# Temperature Column
temperature = np.array(data['Temp'])

# Analysis
print("\n===== WEATHER ANALYSIS REPORT =====")

print("Maximum Temperature :", np.max(temperature))
print("Minimum Temperature :", np.min(temperature))
print("Average Temperature :", np.mean(temperature))
print("Median Temperature :", np.median(temperature))
print("Standard Deviation :", np.std(temperature))

# Hottest and Coldest Day
hottest_day = np.argmax(temperature)
coldest_day = np.argmin(temperature)

print("Hottest Day Index :", hottest_day)
print("Coldest Day Index :", coldest_day)

# Moving Average
window = 30
moving_avg = np.convolve(
    temperature,
    np.ones(window)/window,
    mode='valid'
)

# Visualization
plt.figure(figsize=(12,6))
plt.plot(temperature, label='Temperature')
plt.plot(range(window-1, len(temperature)),
         moving_avg,
         label='30-Day Moving Average')
plt.title("Daily Temperature Analysis")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.legend()
plt.grid(True)
plt.show()
```

###Problem Statement:
Analyze employee salary records to understand salary distribution.

Tasks:

Calculate department-wise average salary.
Find highest and lowest-paid employees.
Compare salaries by gender.
Analyze salary trends.
Generate salary reports.

```python
import pandas as pd
import numpy as np

# Creating a sample dataset since 'employee_salary.csv' is missing
data = {
    'Employee_ID': range(1, 11),
    'Department': ['IT', 'HR', 'IT', 'Sales', 'HR', 'Sales',
                   'IT', 'HR', 'Sales', 'IT'],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male', 'Female',
               'Male', 'Female', 'Male', 'Female'],
    'Salary': [75000, 62000, 80000, 55000, 60000, 58000, 82000,
               65000, 52000, 79000],
    'Year': [2021, 2021, 2022, 2022, 2021, 2023, 2023, 2022, 2021, 2023]
}
df = pd.DataFrame(data)

print("Employee Salary Data")
print(df)

# 1. Department-wise Average Salary
print("\nDepartment-wise Average Salary")
dept_avg = df.groupby("Department")["Salary"].mean()
print(dept_avg)

# 2. Highest and Lowest Paid Employees
highest_paid = df.loc[df["Salary"].idxmax()]
lowest_paid = df.loc[df["Salary"].idxmin()]

print("\nHighest Paid Employee")
print(highest_paid)

print("\nLowest Paid Employee")
print(lowest_paid)

# 3. Salary Comparison by Gender
print("\nAverage Salary by Gender")
gender_salary = df.groupby("Gender")["Salary"].mean()
print(gender_salary)

# 4. Salary Trends by Year
print("\nSalary Trend by Year")
salary_trend = df.groupby("Year")["Salary"].mean()
print(salary_trend)

# 5. Salary Report
print("\nSalary Report")
report = df.groupby("Department").agg(
    Employee_Count=("Employee_ID", "count"),
    Average_Salary=("Salary", "mean"),
    Maximum_Salary=("Salary", "max"),
    Minimum_Salary=("Salary", "min")
)

print(report)

# Save report to CSV
report.to_csv("salary_report.csv")

print("\nSalary report saved successfully!")
```

```python
from sklearn.model_selection import train_test_split

X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X_cls, y_cls,
    test_size=0.2,
    random_state=42,
    stratify=y_cls # Stratify to maintain class distribution in splits
)

print("Shape of X_train_cls:", X_train_cls.shape)
print("Shape of y_train_cls:", y_train_cls.shape)
print("Shape of X_test_cls:", X_test_cls.shape)
print("Shape of y_test_cls:", y_test_cls.shape)
```

```python
from sklearn.linear_model import LogisticRegression

# Initialize the Logistic Regression model
# Set max_iter to a higher value to ensure convergence
logistic_model = LogisticRegression(random_state=42, max_iter=1000)

# Train the model
logistic_model.fit(X_train_cls, y_train_cls)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns

# Make predictions on the test set
y_pred_cls = logistic_model.predict(X_test_cls)

# Calculate classification metrics
accuracy = accuracy_score(y_test_cls, y_pred_cls)
precision = precision_score(y_test_cls, y_pred_cls)
recall = recall_score(y_test_cls, y_pred_cls)
f1 = f1_score(y_test_cls, y_pred_cls)

print("\nLogistic Regression Model Evaluation:")
print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-Score:  {f1:.4f}")

# Display Confusion Matrix
cm = confusion_matrix(y_test_cls, y_pred_cls)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted Low', 'Predicted High'],
            yticklabels=['Actual Low', 'Actual High'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
